```python
#Setting up libraries needed for the project
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
#Creating the initial DataFrame from a CSV file
customers = pd.read_csv('Ecommerce Customers')
```

```python
#Finding all the column names of the initial DataFrame
customers.columns
```

```python
#Setting y (the thing we want predicted) to be "Yearly Amount Spent"
y = customers['Yearly Amount Spent']
```

```python
#Creating a new DataFrame of all the other numerical columns to be used to train the model
X = customers[['Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]
```

```python
#Seperating the data into testing and training splits
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
#Creating an instance of LinearRegression()
lm = LinearRegression()
```

```python
#Using the training split to train the model
lm.fit(X_train,y_train)
```

```python
#Checking the coefficients made by the trained mdoel
lm.coef_
```

```python
#Using the testing data on the trained model to check accuracy
predictions = lm.predict(X_test)
```

```python
#Using a scatterplot to see if the predicted values somewhat match the actual values
plt.scatter(y_test,predictions)
plt.xlabel('Y Test(True Values)')
plt.ylabel('Predicted Values')
```

```python
#Evaluating the model using sklearn
from sklearn import metrics
print('MAE ', metrics.mean_absolute_error(y_test,predictions))
print('MSE ', metrics.mean_squared_error(y_test,predictions))
print('RMSE ', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
#More evaluation
print('Explained Variance Score ',metrics.explained_variance_score(y_test,predictions))
```

```python
#Checking residuals to make sure the graph is close to a bell curve
sns.distplot((y_test-predictions),bins=50)
```
